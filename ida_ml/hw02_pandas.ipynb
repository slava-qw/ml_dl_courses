#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
from io import BytesIO
from zipfile import ZipFile
import requests

url = "https://github.com/hse-ds/iad-intro-ds/blob/master/2023/homeworks/Data.zip?raw=true"
content = requests.get(url)
zip_data = ZipFile(BytesIO(content.content))

csv_files = [f for f in zip_data.namelist() if f.endswith(".csv")]
df_all = pd.DataFrame()
for i in range(10):
    df_info = pd.read_csv(zip_data.open(csv_files[i]))
    df_marks = pd.read_csv(zip_data.open(csv_files[10 + i]))
    df_merged = df_info.join(df_marks.set_index('index'), on='index')
    df_all = df_all.append(df_merged, ignore_index=True)

df_all.sample(10)
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
from collections import defaultdict
from pprint import pprint

start_time1 = time.time()

ans1 = defaultdict(lambda x: float)
d = {'group A': [0, 0],
     'group B': [0, 0],
     'group C': [0, 0],
     'group D': [0, 0],
     'group E': [0, 0]
     }
for row in df_all.values:
    d[row[1]][0] += row[-3]
    d[row[1]][1] += 1

for group, data in d.items():
    ans1[group] = data[0] / data[1]

ex_time1 = time.time() - start_time1


start_time2 = time.time()
ans2 = defaultdict(lambda x: float)
for degree in df_all['education'].unique():
    min_score = 101
    for val in df_all[df_all['education'] == degree]['writing score'].values:
        if val < min_score:
            min_score = val
    ans2[degree] = min_score
        
        

ex_time2 = time.time() - start_time2



pprint(ans1)
print('\n', f'time1: {ex_time1}')
print()
pprint(ans2)
print('\n', f'time2: {ex_time2}')
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице `df_all` под названием `Math_PassStatus` и запишите в него `F`, если студент не сдал экзамен по математике (балл за экзамен < `passmark`), и `P` иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
groups = []
for name in ['math score', 'reading score', 'writing score']:
    sub = name.split()[0].title()
    df_all[f'{sub}_PassStatus'] = 'P'
    df_all.loc[df_all[name] < passmark, f'{sub}_PassStatus'] = 'F'
    
    groups.append(df_all.groupby([f'{sub}_PassStatus'])[f'{sub}_PassStatus'].count())

pd.concat(groups, axis=1)
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец `Grade` и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
        sw = np.sum(np.linspace(50, 90, 5) < average_mark)  
        return chr(70 - sw)

df_all['Grade'] = df_all[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)
df_all.groupby(['Grade'])['Grade'].count()
```
